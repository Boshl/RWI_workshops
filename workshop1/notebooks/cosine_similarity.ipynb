# Cosine Similarity
Basierend auf folgender Quelle: https://studymachinelearning.com/cosine-similarity-text-similarity-metric/

```python
with open('../misc/cats.txt') as f:
    cats = f.read()

print(cats)
```

```python
with open('../misc/dogs.txt') as f:
    dogs = f.read()

print(dogs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

data = [cats, dogs]
count_vectorizer = CountVectorizer()
vector_matrix = count_vectorizer.fit_transform(data)
tokens = count_vectorizer.get_feature_names()
tokens
```

```python
import pandas as pd

def create_dataframe(matrix, tokens):
    doc_names = ['cats', 'dogs']
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)

create_dataframe(vector_matrix.toarray(),tokens)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_dataframe(cosine_similarity_matrix,['cats','dogs'])
```

## Tf-idf Vector

```python
from sklearn.feature_extraction.text import TfidfVectorizer

data = [cats, dogs]
count_vectorizer = TfidfVectorizer()
vector_matrix = count_vectorizer.fit_transform(data)
tokens = count_vectorizer.get_feature_names()

create_dataframe(vector_matrix.toarray(),tokens)
```
